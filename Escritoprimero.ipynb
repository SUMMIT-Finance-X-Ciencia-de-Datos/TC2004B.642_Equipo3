# Cost - Report

En esta última sección se presenta un reporte de costos que analiza la relación entre los costos reales y los presupuestos asignados, así como su conexión con las diferentes cuentas que se manejan. Este análisis es clave para entender cómo se están utilizando los recursos y qué tan alineados están los gastos con lo planeado.
```python
df_data = df_Data.rename(columns={"PARTIDA PRESUPUESTAL": "ACCT"})
df_data
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)
# Agrupa el DataFrame por la columna "ACCT" y suma los valores de SUBTOTAL por cuenta.
df_data = df_data.groupby("ACCT", as_index=False)[["SUBTOTAL"]].sum()
df_data
```

```python
df_CostReport1=pd.merge(budget,df_data, on="ACCT", how="outer")
df_CostReport2= df_CostReport1.rename(columns = {"SUBTOTAL": "COST TO DATE"})
df_CostReport2
```

```python
df_CostReport3 = df_CostReport2.groupby("ACCT", as_index=False).agg({
    "TIPO": "first",
    "REF": "first",
    "Etapa": "first",
    "BUDGET": "first",
    "COST TO DATE": "sum"
})
df_CostReport3
```

```python
df_CostReport3["VARIANCE"] = df_CostReport3["BUDGET"] - df_CostReport3["COST TO DATE"]
df_CostReport4 = df_CostReport3.groupby("ACCT", as_index=False).agg({
    "TIPO": "first",
    "REF": "first",
    "Etapa": "first",
    "BUDGET": "first",
    "COST TO DATE": "sum"
})

# Agregar la columna de diferencia
df_CostReport4["VARIANCE"] = df_CostReport4["BUDGET"] - df_CostReport4["COST TO DATE"]

# Mostrar el resultado
df_CostReport4.head(15)
```

## **

```python
df = pd.read_excel('BASE.xlsx', sheet_name=None)
comprometidos = df['Comprometidos']
comprometidos = comprometidos.drop(['CONCEPTO', 'MONTO', 'PAGO'], axis=1)
comprometidos = comprometidos.rename(columns={"CUENTA": "ACCT", "NVO MONTO": "COMMITMENTS"})

# Asegurar tipo de dato
df_CostReport4["ACCT"] = df_CostReport4["ACCT"].astype(str)
comprometidos["ACCT"] = comprometidos["ACCT"].astype(str)

CostReport = pd.merge(df_CostReport4, comprometidos, on="ACCT", how="left")
CostReport["COMMITMENTS"] = CostReport["COMMITMENTS"].fillna(0)

# Filtrar solo las cuentas de detalle
detalle_df = CostReport[CostReport["REF"] == "Detalle"].copy()

# Crear subcuentas 
sub_df = detalle_df.copy()
sub_df["SUB_ACCT"] = sub_df["ACCT"].str[:4]

# Agrupar por subcuenta, usando el TIPO más frecuente de cada grupo
sub_sums = sub_df.groupby("SUB_ACCT", as_index=False).agg({
    "BUDGET": "sum",
    "COST TO DATE": "sum",
    "COMMITMENTS": "sum",
    "TIPO": lambda x: x.mode()[0] if not x.mode().empty else x.iloc[0]
})
sub_sums["REF"] = "Sub cuenta"
sub_sums = sub_sums.rename(columns={"SUB_ACCT": "ACCT"})

# Crear cuentas generales 
sub_sums["GEN_ACCT"] = sub_sums["ACCT"].str[:2] + "00"

# Agrupar por cuenta general
gen_sums = sub_sums.groupby("GEN_ACCT", as_index=False).agg({
    "BUDGET": "sum",
    "COST TO DATE": "sum",
    "COMMITMENTS": "sum",
    "TIPO": lambda x: x.mode()[0] if not x.mode().empty else x.iloc[0]
})
gen_sums["REF"] = "Cuenta General"
gen_sums = gen_sums.rename(columns={"GEN_ACCT": "ACCT"})

final_cost_report = pd.concat([gen_sums, sub_sums, detalle_df], ignore_index=True)
final_cost_report[["BUDGET", "COST TO DATE", "COMMITMENTS"]] = final_cost_report[["BUDGET", "COST TO DATE", "COMMITMENTS"]].fillna(0)
final_cost_report["VARIANCE"] = final_cost_report["BUDGET"] - (final_cost_report["COST TO DATE"] + final_cost_report["COMMITMENTS"])
pd.set_option('display.float_format', '{:,.2f}'.format)
final_cost_report.to_csv("final_cost_report.csv", index=False)
```

```python
# Crea una app interactiva en Dash para visualizar el TOTAL CTD por tipo con un gráfico circular y selector
df = final_cost_report[final_cost_report["REF"] == "Cuenta General"].copy()
resumen = df.groupby("TIPO", as_index=False)["TOTAL CTD"].sum() # Agrupa por tipo de cuenta y suma el TOTAL CTD
resumen = resumen.sort_values("TOTAL CTD", ascending=False)

app = dash.Dash(__name__)
# Define la estructura visual de la app (título, checklist, gráfico, total mostrado)
app.layout = html.Div([
    html.H2('Distribución del TOTAL (CTD) por TIPO'),
    dcc.Checklist( id='tipo-selector', options=[{"label": tipo, "value": tipo} for tipo in resumen["TIPO"]], value=list(resumen["TIPO"]),  # Todos seleccionados por default
        inline=True
    ),
    dcc.Graph(id='pie-cost', config={"displayModeBar": False}),
    html.Div(id='total-box', style={
        'fontSize': '26px', 'fontWeight': 'bold',  'marginTop': '20px', 'textAlign': 'center', 'backgroundColor': '#f4f4f4', 'borderRadius': '15px', 'padding': '15px', 'width': '50%',  'margin': '0 auto'
    })
])

@app.callback(
    [Output('pie-cost', 'figure'), Output('total-box', 'children')], [Input('tipo-selector', 'value')]
)
def update_pie(selected_tipos):    # Filtra el resumen para mostrar solo los tipos seleccionados
    resumen_filtrado = resumen[resumen["TIPO"].isin(selected_tipos)]
    total_actual = resumen_filtrado["TOTAL CTD"].sum()

    fig = px.pie(
        resumen_filtrado, values="TOTAL CTD",  names="TIPO",  hole=0.5, title="<b>Porcentaje de uso por TIPO </b>"
    )
    fig.update_traces(
        textposition='inside',
        textinfo='percent+label',
        pull=[0.02]*len(resumen_filtrado),
        marker=dict(line=dict(color='white', width=2))
    )
    fig.update_layout(
        showlegend=True, title_font_size=22,legend_title_text='Tipo de Cuenta',legend=dict(font=dict(size=14)),margin=dict(t=80, b=40), annotations=[
            dict(
                text="",font_size=18,showarrow=False,
                x=0.5, y=0.5,
                xref="paper", yref="paper",
                align='center'
            )
        ]
    )
    return fig, f"TOTAL ACTUAL: ${total_actual:,.0f}"

if __name__ == '__main__':
    app.run(debug=True, port=8051)
```

Se muestran las 10 cuentas con mayor ahorro con respecto al presupuesto asignado. El objetivo principal es visualizar cuáles han tenido una ejecución presupuestal más baja, lo que puede indicar una mala gestión de recursos, un subejercicio o incluso una sobreestimación del presupuesto original.

El gráfico 7 facilita la visualización y comparación del uso del presupuesto entre cuentas globales. Al ser interactiva, permite identificar qué cuentas representan los mayores porcentajes del gasto total.

El usuario puede seleccionar cuentas específicas para analizar su peso relativo dentro del total, lo que resulta útil para detectar concentraciones de recursos, lo que también ayuda a tomar decisiones más informadas sobre la distribución presupuestal y posibles ajustes financieros.

```python

final_cost_report["ACCT"] = final_cost_report["ACCT"].astype(str)

# Clasificación según cuenta
def clasificar_cuenta(c):
    if "-" in c:
        return "Detalle"
    elif c.isdigit() and len(c) == 4:
        return "Cuenta general" if int(c) % 100 == 0 else "Subcuenta"
    else:
        return "Otro"

final_cost_report["NIVEL_CUENTA"] = final_cost_report["ACCT"].apply(clasificar_cuenta)

# Eficiencia: ahorro vs sobrecosto
final_cost_report["EFICIENCIA"] = np.where(final_cost_report["VARIANCE"] >= 0, "Ahorro", "Sobrecosto")

# Limpiar y preparar
df_bubble = final_cost_report.dropna(subset=["BUDGET", "COST TO DATE", "VARIANCE"]).copy()

# Eliminar las cuentas 6137 y 6137-003
df_bubble = df_bubble[~df_bubble["ACCT"].isin(["6137", "6137-003"])]


df_bubble = df_bubble.rename(columns={"VARIANCE": "DIFERENCIA"})
df_bubble["DIFERENCIA"] = df_bubble["DIFERENCIA"].abs()

# Gráfico inicial
nivel_inicial = "Detalle"
df_inicial = df_bubble[df_bubble["NIVEL_CUENTA"] == nivel_inicial]

# Función para generar datos por tipo de cuenta
def generar_args(tipo):
    if tipo == "Todos":
        df_tipo = df_bubble
    else:
        df_tipo = df_bubble[df_bubble["NIVEL_CUENTA"] == tipo]
    return [
        {
            "x": [df_tipo["BUDGET"]],"y": [df_tipo["COST TO DATE"]],
            "marker.size": [df_tipo["DIFERENCIA"]],"hovertext": [ 
                df_tipo.apply(
                    lambda row: f"Cuenta: {row['ACCT']}<br>Diferencia: ${row['DIFERENCIA']:,.2f}<br>Eficiencia: {row['EFICIENCIA']}",
                    axis=1
                )
            ], "marker.color": [df_tipo["EFICIENCIA"].map({"Ahorro": "green", "Sobrecosto": "red"})]
        }
    ]

# Crear gráfico
fig = px.scatter(df_inicial,x="BUDGET", y="COST TO DATE",
    size="DIFERENCIA", color="EFICIENCIA", hover_name="ACCT",
    category_orders={"EFICIENCIA": ["Ahorro", "Sobrecosto"]},
    color_discrete_map={"Ahorro": "green", "Sobrecosto": "red"},title=f"<b>Gráfico 6. Impacto Financiero por Cuenta </b>",
    labels={"BUDGET": "BUDGET ($)", "COST TO DATE": "COST TO DATE ($)","EFICIENCIA": "Clasificación", "DIFERENCIA": "DIFFERENCE ($)"}, size_max=60, height=600
)

# Menú interactivo
fig.update_layout(
    title={'x': 0.5},
    title_font_size=24,
    updatemenus=[
        {
            "buttons": [
                {"method": "restyle", "label": "Detalle", "args": generar_args("Detalle")},
                {"method": "restyle", "label": "Subcuenta", "args": generar_args("Subcuenta")},
                {"method": "restyle", "label": "Cuenta general", "args": generar_args("Cuenta general")},
                {"method": "restyle", "label": "Todos", "args": generar_args("Todos")},
            ],
            "direction": "down", "showactive": True, "x": 1.05,   "xanchor": "left", "y": 1.15,"yanchor": "top"
        }
    ],
    legend=dict(title="Eficiencia")
)

fig.show()

```

El gráfico 8, el cual es una gráfica tipo scatterplot permite analizar visualmente la relación entre el presupuesto asignado y el gasto real (Cost to Date) por cuenta. Cada punto representa una cuenta específica, donde el eje X indica el presupuesto y el eje Y muestra el costo real. El tamaño de cada burbuja refleja la magnitud de la variación entre ambos valores, mientras que el color permite identificar rápidamente el tipo de resultado: verde para cuentas con ahorro y rojo para cuentas con sobrecosto. Además, al ser interactiva, permite consultar los detalles de cada cuenta con solo pasar el cursor sobre las burbujas, haciendo de esta herramienta una forma intuitiva y poderosa de explorar la información financiera.
